```python
import nltk 
from nltk.corpus import stopwords
```

```python
def text_cleaning(text):
    no_punctuation = [char for char in text if char not in string.punctuation]
    no_punctuation = ''.join(no_punctuation)
    return [word for word in no_punctuation.split() if word.lower() 
            not in stopwords.words('english')]
```

```python
data['Policy Statement'].apply(text_cleaning)
```

```python
data["Policy Statement"].head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
bag_of_words = CountVectorizer(analyzer = text_cleaning).fit(data['Policy Statement'])
```

```python
print(bag_of_words)
```

```python
print(bag_of_words.vocabulary_)
```

```python
message_bow = bag_of_words.transform(data['Policy Statement'])
```

```python
print(message_bow)
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
tfidf_transformer = TfidfTransformer().fit(message_bow)
```

```python
message_tfidf = tfidf_transformer.transform(message_bow)
```

```python
print(message_tfidf.shape)
print(message_tfidf)
```

```python
X = message_tfidf
Y = data['Classification']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size = 0.2)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn import neighbors
p_d = MultinomialNB().fit(X_train,Y_train)
```

```python
predict = p_d.predict(X_test)
print(predict)
```

```python
expected = Y_test

from sklearn import metrics
print(metrics.classification_report(expected, predict))
print(metrics.confusion_matrix(expected, predict))
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_test,predict)
```
